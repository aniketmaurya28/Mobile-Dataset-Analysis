## Part 2: Data cleaning & type conversion

6. String processing & astype

```python
df['MRP'] = df['MRP'].astype(str).str.replace('₹','').str.replace(',','').replace('NIL', np.nan).astype(float)                                                                            # By Aniket Maurya
df
```

```python
df['Selling_Price']= df['Selling_Price'].str.replace('₹','').str.replace(',','').replace('NIL', np.nan).astype(float)
df
```

```python
df['Rating'] = df['Rating'].replace(' Ratings','',regex=True).replace(',','',regex=True).astype(int)                                                                                                                                                                                    #By Aniket Maurya
```

```python
df['Reviews'] = df['Reviews'].replace(' Reviews','',regex= True).replace(',','',regex=True).astype(int)
```

```python
df
```

7. Handling missing values

```python
df['Camera'].isnull().sum()
```

```python
df.fillna({'Camera': 'Not specified'},inplace=True)
```

```python
df['Stars'].dropna(inplace=True)
```

8. Duplicates

```python
df.drop_duplicates(inplace=True)
```

9. Replace

```python
df['RAM (GB)'].replace('NIL','np.nan',inplace=True)
```

## Part 4: Advance Series and String operarions

14. value_count & unique

```python
df['Brand'] = df['Product Name'].str.split().str[0]
```

15. map

```python
conditions = [(df['Stars'] <= 3), (df['Stars'] <= 4)]         # By AM
df['Price_Category'] = np.select(conditions, ['Budget', 'Mid-Range'], default='Premium')
df
```

16. apply & assign

```python
# Discount (Aniket Maurya) Percentage: $((MRP -Selling\_Price) / MRP) * 100$.

df['Discount_Percentage'] = ((df['MRP'] - df['Selling_Price']) / df['MRP']) * 100
df
```

## Part 5: Grouping & Aggregation

18. groupby & agg

19. Transform

```python
df['Average_Brand_Price'] = df.groupby('Brand')['Selling_Price'].transform(lambda x:x.mean())           # By AM
df
```

20. Pivot table

```python
pivot_table = pd.pivot_table(df, values='Stars', index='Brand', columns='RAM (GB)', aggfunc='mean')
pivot_table
```

## Part 7: Time Series & Numpy

25. Date Creation

```python
df['Launch_Date'] = pd.date_range('2023-01-01', periods=len(df), freq='D')
df
```

26. Date Accessors

```python
df['Year'] = df['Launch_Date'].dt.year
df['Month_Name'] = df['Launch_Date'].dt.month_name()
df['Day_of_Week'] = df['Launch_Date'].dt.day_name()
df
```

27. Time periods

```python
df['Launch_Period'] = df['Launch_Date'].dt.to_period('M')
```

28. Numpy integration

```python
prices_array = df['Selling_Price'].to_numpy()
np.percentile(prices_array, 75)
```
